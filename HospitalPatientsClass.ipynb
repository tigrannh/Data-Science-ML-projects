```python
X_data['Patient_Smoker'].unique()
```

```python
def changeDataset(X):
    for i in range(1,7):
        dx = f'dx{i}'
        X[dx] = X.Treated_with_drugs.fillna('0').str.contains(dx).astype(int)
    X.drop('Treated_with_drugs', axis=1, inplace=True)
    X.drop('Patient_ID', axis=1, inplace=True)
    X.drop('Number_of_prev_cond', axis=1, inplace=True)
    X.drop('Patient_mental_condition', axis=1, inplace=True)
    X['Patient_Smoker'] = X['Patient_Smoker'].str.strip().str.lower().replace('yess','yes')
    
    for col in X.columns:
        if X[col].dtype == 'object':
            X[col] = X[col].astype('category')

    
```

```python
changeDataset(X_data)
X_data
```

```python
X_data['Patient_Smoker'].unique()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.15)
X_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder


new_data = data.dropna()
changeDataset(new_data)
new_train, new_test, new_y_train, new_y_test = train_test_split(new_data.drop('Survived_1_year', axis=1), new_data['Survived_1_year'], test_size=0.15)

lb_make = LabelEncoder()
new_train.loc[:,'Patient_Smoker'] = lb_make.fit_transform(new_train['Patient_Smoker'])
new_train.loc[:,'Patient_Rural_Urban'] = lb_make.fit_transform(new_train['Patient_Rural_Urban'])
new_test.loc[:,'Patient_Smoker'] = lb_make.fit_transform(new_test['Patient_Smoker'])
new_test.loc[:,'Patient_Rural_Urban'] = lb_make.fit_transform(new_test['Patient_Rural_Urban'])
```

```python
new_train
```

```python
from sklearn.svm import SVC
```

```python
svm_classifier = SVC()
svm_classifier.fit(new_train, new_y_train)

print('train accuracy: ', accuracy_score(svm_classifier.predict(new_train), new_y_train))
print('test accuracy: ', accuracy_score(svm_classifier.predict(new_test), new_y_test))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': list(np.linspace(1, 1000, 10)),
              'gamma': [0.3, 0.1, 0.03, 0.01],
              'kernel': ['rbf', 'poly', 'linear']}
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
grid.fit(new_train, new_y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier

```

```python
dtc = DecisionTreeClassifier()
dtc.fit(new_train, new_y_train)

print('train accuracy: ', accuracy_score(dtc.predict(new_train), new_y_train))
print('test accuracy: ', accuracy_score(dtc.predict(new_test), new_y_test))
```
